## import library

```python
import cv2
```

## read the image

```python
image = cv2.imread("C:/Users/tusha/Pictures/hp.jpg")
```

## to display the image

```python
cv2.imshow("Harry Potter",image)
cv2.waitKey(0)
```

## convert the original image into greyscale image

```python
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
cv2.imshow("greyscale image",gray_image)
cv2.waitKey(0)
```

## invert the greyscale image

```python
inverted_image = 255 - gray_image
```

```python
cv2.imshow("Inverted image",inverted_image)
cv2.waitKey(0)
```

## blur the image using Gaussian Function in OpenCV

```python
blur = cv2.GaussianBlur(inverted_image,(21,21),0)
```

```python
cv2.imshow("Blurred image", blur)
cv2.waitKey(0)
```

## invert the blurred image

```python
inverted_blur = 255 - blur
```

```python
cv2.imshow("Inverted blur image", inverted_blur)
cv2.waitKey(0)
```

## pencil sketch

```python
pencil_sketch = cv2.divide(gray_image,inverted_blur,scale = 256 )
```

```python
cv2.imshow("Pencil Sketch",pencil_sketch)
cv2.waitKey(0)
```

## original image vs pencil sketch

```python
cv2.imshow("Original Image",image)
cv2.imshow("Pencil Sketch",pencil_sketch)
cv2.waitKey(0)
```
